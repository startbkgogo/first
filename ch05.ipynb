```python
import gdown
gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
# scatter 그래프 그리기
import matplotlib.pyplot as plt
# 산점도 그래프 출력
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# 그래프 크기 바꾸기 :  figsize(피규어 크기) 매개 변수
# figure : 도표의 크기 세팅, figsize=(x축의 크기, y축의 크기) : 단위 인치 입니다.
plt.figure(figsize=(9, 6)) # 너비가 9인치, 높이가 6인치인 피켜 객체를 생성합니다.
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# 기본 그래프의 크기 확인
print(plt.rcParams['figure.figsize'])
```

```python
# 기본 해상도 확인
print(plt.rcParams['figure.dpi'])
```

```python
plt.figure(figsize=(900/72, 600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
# plt.xlabel('도서권수')
# plt.ylabel('대출건수')
plt.show()
```

```python
# 고랩의 설정값을 바꾼다. - 적용된다. : 출력할 때 경계선에 붙어는 것은 불력범위를 벗어나서 출력이 안된다. 출력 여백 추가해서
%config InlineBackend.print_figure_kwargs = {'bbox_inches': None}
plt.figure(figsize=(900/72, 600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
%config InlineBackend.print_figure_kwargs = {'bbox_inches': 'tight'}
```

```python
# 그래프 크기 바꾸기 : dpi 매개 변수
plt.figure(dpi=144)
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# DPI 기본값 바꾸기
plt.rcParams['figure.dpi'] = 100
```
